### Cleaning and making data feasible table wise

```python
# From " content " cleaning and removing data.

contents.drop(contents.columns[[0,3]], axis = 1, inplace = True)
contents.head()
```

```python
# From " profile " cleaning and removing data.

profile.drop(profile.columns[0], axis = 1, inplace = True)
profile.head()
```

### Joining the datasets using its keys 

```python
# Using outer join to consider all columns for use
```

```python
Join_1 = pd.merge(profile, contents, how='outer', on='User ID')
Join_1.head()
```

```python
Join_1.drop(Join_1.columns[[4]], axis = 1, inplace = True)
Join_1.head()
```

```python
Join_2 = pd.merge(Join_1, reactions, how='outer', on='Content ID')
Join_2.head()
```

```python
Join_3 = pd.merge(Join_2, reaction_types , how='outer', on='Type')
Join_3.head()
```

### removing null values

```python
df1 =Join_3.dropna()
```

```python
df1.head()
```

```python
df1.drop(df1.columns[[5,6,9]], axis = 1, inplace = True)
```
